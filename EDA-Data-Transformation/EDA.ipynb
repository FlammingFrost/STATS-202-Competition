```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('../data/rawdata/train.csv')
df_test = pd.read_csv('../data/rawdata/test.csv')
## Remane test "ZIP" to "Zip"
df_test = df_test.rename(columns={'ZIP': 'Zip'})

## delete the potential error data - 5th row
df = df.drop([4], axis=0)
## Drop the column 'Id' from the data
# df = df.drop(['ID'], axis=1)
# df_test = df_test.drop(['ID'], axis=1)
```

```python
n_predictors = len(df.columns) - 1
df.columns
```

```python
# Plot the data
fig, axs =  plt.subplots(2, 4, figsize=(20, 10), dpi = 400)
for predictor in range(n_predictors):
    i, j = predictor // 4, predictor % 4
    axs[i, j].scatter(df.iloc[:, predictor], df.iloc[:, -1])
    axs[i, j].set_xlabel(df.columns[predictor])
    axs[i, j].set_ylabel(df.columns[-1])

plt.tight_layout()
plt.show()
```
